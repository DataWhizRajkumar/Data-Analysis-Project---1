# 1. Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

# 2. Prepare Data

```python
data = pd.read_csv('Advertising Sales Channel Prediction.csv')

# Seperating features (X) and target variable (Y)
X = data[['TV']]
Y = data['Sales']

# Splitting the data into training and testing sets (70% training, 30% testing)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 42)
```

# 3. Fit the Model 

```python
# Creating a LinearRegression model
model = LinearRegression()

# Fitting the model to the training data
model.fit(X_train, Y_train)
```

# 4. Make Predictions

```python
# Making predictions for the testing data
predicted_Y = model.predict(X_test)
print("Predicted Y (Testing Data): ", predicted_Y)
```

# 5. Visualize the Results

```python
# Plotting the data points
plt.scatter(X.values.reshape(-1), Y.values.reshape(-1))

#Plotting the regression line
plt.plot(X, model.predict(X), color = 'red', label = 'Regression Line')

plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('TV')
plt.legend()
plt.show()
```

```python
# Plotting the trained data points
plt.scatter(X_train.values.reshape(-1), Y_train.values.reshape(-1))

#Plotting the regression line
plt.plot(X_train, model.predict(X_train), color = 'red', label = 'Regression Line')

plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Trained data')
plt.legend()
plt.show()
```

```python
# Plotting the test data points
plt.scatter(X_test.values.reshape(-1), Y_test.values.reshape(-1))

#Plotting the regression line
plt.plot(X, model.predict(X), color = 'red', label = 'Regression Line')

plt.xlabel('X (Independent Variable)')
plt.ylabel('Y (Dependent Variable)')
plt.title('Test data')
plt.legend()
plt.show()
```
